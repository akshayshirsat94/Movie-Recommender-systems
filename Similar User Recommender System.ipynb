- ## Similar User Recommender System :

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
column_names =[ 'user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv("u.data", sep = '\t', names = column_names)
df.head()
```

```python
movie_titles = pd.read_csv("Movie_Id_Titles")
movie_titles.head()
```

```python
df = pd.merge(movie_titles,df,on='item_id')
df.head()
```

```python
z1 = df['user_id'].value_counts()
print(z1)
```

```python
userRatings = df.pivot_table(index=['title'],columns=['user_id'],values='rating')
userRatings
```

```python
userRatings.shape
```

```python
print("Before: ",userRatings.shape)
userRatings = userRatings.dropna(thresh=10, axis=1).fillna(0,axis=1)
#userRatings.fillna(0, inplace=True)
print("After: ",userRatings.shape)
```

```python
corrMatrix = userRatings.corr(method='pearson')
corrMatrix.head()
```

```python
def get_similar(user_id,rating):
    similar_ratings = corrMatrix[user_id]*(rating-2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    #print(type(similar_ratings))
    return similar_ratings
```

```python
user = [(50,3)]
similar_user = pd.DataFrame()
for user_id,rating in user:
    similar_user = similar_user.append(get_similar(user_id,rating),ignore_index = True)

similar_user.head(10)
a = similar_user.sum().sort_values(ascending=False)
a
```

```python
b = pd.DataFrame(a, columns =['corr'])
b['user_id'] = b.index
b
```
